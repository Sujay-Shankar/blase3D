```python
import torch
import altair as alt

from gollum.phoenix import PHOENIXSpectrum

alt.data_transformers.enable('vegafusion')
```

```python
specs = [PHOENIXSpectrum(teff=teff, logg=4, Z=0, wl_lo=8530, wl_hi=8550).normalize().instrumental_broaden(40000) for teff in range(3000, 12001, 200)]
colors = [f'#{n:06x}' for n in range(100000, 100100, 5)]
```

```python
alt.layer(*[alt.Chart(spec.to_pandas(), width=1800, height=800).mark_line().encode(x='wavelength', y='flux', color=alt.value(color)).interactive() for spec, color in zip(specs, colors)])
```
